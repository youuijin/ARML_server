```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv("./logs/AAresult_csv/"+"at_standard_Linf_6_0910133124.csv")
```

```python
plt.figure(figsize=(10,7))
plt.title("nat_standard_Linf_6_0910133124.csv")

colors = plt.cm.jet(np.linspace(0, 1, len(data)))

for idx, row in data.iterrows():
    # if row['model'].startswith('R-MAML-AT_PGD-Linf'):
    #     continue
    if row['model'].split('_')[1]=='aRUB':
        marker = 'o'
    elif row['model'].split('_')[1]=='PGD-Linf':
        marker = '^'
    else:
        marker = 'x'
    plt.scatter(row['SA'], row['RA'], label=row['model'], color=colors[idx], marker=marker)
    # plt.scatter(row['SA'], row['RA'], label=row['model'])

plt.xlabel('SA')
plt.ylabel('RA')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.tight_layout()

plt.ylim(0, 0.01)

plt.show()
```
